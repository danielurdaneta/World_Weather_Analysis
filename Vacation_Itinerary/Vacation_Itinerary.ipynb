## Deliverable 3. Create a Travel Itinerary Map.

```python
# Dependencies and Setup
import pandas as pd
import requests
import gmaps

# Import API key
from config import g_key

# Configure gmaps
gmaps.configure(api_key=g_key)
```

```python
# 1. Read the WeatherPy_vacation.csv into a DataFrame.
vacation_df = pd.read_csv("vacation_search/weatherpy_vacations.csv")
vacation_df
```

```python
# 2. Using the template add the city name, the country code, the weather description and maximum temperature for the city.
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City Name</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Weather Description</dt><dd>{Weather Description}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
</dl>
"""

# 3a. Get the data from each row and add it to the formatting template and store the data in a list.
vacation_info = [info_box_template.format(**row) for index, row in vacation_df.iterrows()]

# 3b. Get the latitude and longitude from each row and store in a new DataFrame.
locations = vacation_df[["Lat", "Lng"]]
```

```python
# 4a. Add a marker layer for each city to the map.
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
marker_layer = gmaps.marker_layer(locations, info_box_content=vacation_info)
fig.add_layer(marker_layer)
# 4b. Display the figure
fig
```

```python
# From the map above pick 4 cities and create a vacation itinerary route to travel between the four cities. 
# 5. Create DataFrames for each city by filtering the 'vacation_df' using the loc method. 
# Hint: The starting and ending city should be the same city.

vacation_start = vacation_df.loc[vacation_df["City"] == "Naron"]
vacation_end = vacation_df.loc[vacation_df["City"] == "Naron"]
vacation_stop1 = vacation_df.loc[vacation_df["City"] == "Muros"]
vacation_stop2 = vacation_df.loc[vacation_df["City"] == "Arroyo"] 
vacation_stop3 = vacation_df.loc[vacation_df["City"] == "Bermeo"]

vacation_start
```

```python
# 6. Get the latitude-longitude pairs as tuples fom each city DataFrame using the to_numpy function and list indexing.
start = vacation_start[["Lat","Lng"]].to_numpy(dtype=float)
start_lt_lg = (start[0][0], start[0][1])

end = vacation_end[["Lat","Lng"]].to_numpy(dtype=float)
end_lt_lg = (end[0][0], end[0][1])

stop1 = vacation_stop1[["Lat","Lng"]].to_numpy(dtype=float)
stop_lt_lg1 = (stop1[0][0], stop1[0][1])

stop2 = vacation_stop2[["Lat","Lng"]].to_numpy(dtype=float)
stop_lt_lg2 = stop2[0][0], stop2[0][1]

stop3 = vacation_stop3[["Lat","Lng"]].to_numpy(dtype=float)
stop_lt_lg3 = stop3[0][0], stop3[0][1]
```

```python
# 7. Create a direction layer map using the start and end latitude-longitude pairs,
# and stop1, stop2, and stop3 as the waypoints. The travel_mode should be "DRIVING", "BICYCLING", or "WALKING".
fig = gmaps.figure()
itinerary = gmaps.directions_layer(start_lt_lg, end_lt_lg, waypoints= [stop_lt_lg1,stop_lt_lg2, stop_lt_lg3], travel_mode = "DRIVING")
fig.add_layer(itinerary)


fig
```

```python
# 8. To create a marker layer map between the four cities.
#  Combine the four city DataFrames into one DataFrame using the concat() function.
itinerary_df = pd.concat([vacation_start, vacation_stop1, vacation_stop2, vacation_stop3],ignore_index=True)
itinerary_df
```

```python
# 9 Using the template add city name, the country code, the weather description and maximum temperature for the city. 
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City Name</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Weather Description</dt><dd>{Weather Description}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
</dl>
"""
fig = gmaps.figure()

# 10a Get the data from each row and add it to the formatting template and store the data in a list.
vacation_info = [info_box_template.format(**row) for index, row in itinerary_df.iterrows()]

# 10b. Get the latitude and longitude from each row and store in a new DataFrame.
locations = itinerary_df[["Lat", "Lng"]]

# 11a. Add a marker layer for each city to the map.
marker_layer = gmaps.marker_layer(locations, info_box_content=vacation_info)
fig.add_layer(marker_layer)

# 11b. Display the ffig.add_layer(marker_layer)
fig

```
