```python
#Import dependecies
import pandas as pd
import numpy as np
from citipy import citipy
from config import weather_api_key
import requests
```

```python
#Generate 2000 latitudes and logitudes
lats= np.random.uniform(low= -90.000, high= 90.000, size = 2000)
longs= np.random.uniform(low= -180.000, high= 180.000, size = 2000)

#Generate an array with latitudes and longitudes
lat_lng = zip(lats,longs)
coordinates = list(lat_lng)
```

```python
#Found the nearest city

cities = []
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    if city not in cities:
        cities.append(city)
```

```python
#Generate basic url
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID="
```

```python

#Create an empty list
city_data = []

#Print beginning Data Retrieval
print("Beginning Data Retrieval     ")
print("----------------------------")


for i, city in enumerate(cities):
        
    try:
        #Create URL
        city_url = url + weather_api_key + "&q=" + city.replace(" ", "+")

        #log the URL, record, and set numbers and the city
        print(f"Processing Record {city}")


                # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()

                    # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_description = city_weather["weather"][0]["description"]

                    # Append the city information into city_data list.
        city_data.append({"City": city.title(),"Country": city_country,"Lat": city_lat,"Lng": city_lng,"Max Temp": city_max_temp,"Humidity": city_humidity,"Cloudiness": city_clouds,"Wind Speed": city_wind, "Weather Description": city_description})
    except:
        print("error")
        pass

#Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
#Generate a database
weather_database = pd.DataFrame(city_data)
```

```python
#Save as a csv file

# Create the output file (CSV).
output_data_file = "weather_database/WeatherPy_Database.csv"
# Export the City_Data into a CSV.
weather_database.to_csv(output_data_file, index_label="City_ID")
```
